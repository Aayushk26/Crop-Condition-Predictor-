```python
x=train.drop(['ID','Crop_Damage'],axis=1)
xt=test.drop(['ID'],axis=1)
y=train.Crop_Damage
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from lightgbm import LGBMClassifier
lgbm=LGBMClassifier(eval_metric='multi_error',objective='multiclass')
```

```python
lgbm.fit(x_train,y_train)
```

```python
ypred=lgbm.predict(x_val)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy Score using LGBM:',accuracy_score(y_val,ypred))
from sklearn.metrics import classification_report
print(classification_report(y_val,ypred))
```
